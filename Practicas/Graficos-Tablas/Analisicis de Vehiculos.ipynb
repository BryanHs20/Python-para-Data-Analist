# Importacion de librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importacion del archivo csv
df = pd.read_csv("../../Archivos-Analisis/vehicle_price_prediction.csv")
df
```

```python
# Informacion del DF
df.info()
```

```python
# Estadistica descriptiva
df.describe()
```

```python
# Años con mas ventas
df["year"].value_counts().sort_index().plot(kind="bar")
plt.title("Años con mas ventas")
```

```python
# Modelos + vendidos
df["model"].value_counts().head(10).plot(kind="bar")
plt.title("Vehiculos mas vendidos")
```

```python
sns.regplot(data=df, x="year", y="price")v
```

```python
fig, ax = plt.subplots()
sns.stripplot(data = df, y="make", x ="model", hue="year", jitter=True)
plt.title("Marcas y Modelos mas Vendidos x Año")
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.show()
```

```python
df["fuel_type"].value_counts().sort_index().plot(kind="bar")
plt.title("Tipos de Combustible")
```

```python
# Autos mas viejos
df_more_old = df[df["year"] < 2005].sort_values(by="year").head(10)
df_more_old
```

```python
# Mas caros
df_max_price = df.sort_values(by="price", ascending=False).head(10)
df_max_price
```

```python
sns.regplot(data = df, x="make", y = "price")
plt.xticks(rotation=90)
plt.tick_params(axis='y', which='both', left=False, right=False, labelleft=False)
plt.title("Comparacion de ventas entre marcas")
plt.show()
```
