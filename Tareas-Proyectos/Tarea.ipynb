```python
# Lo primero sera importat todas las librerias de python que vamos a utilizar
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Mediante el uso de gráficos de su preferencia (de acuerdo a los diversos tipos discutidos en este módulo), muestre la relación que guardan la tasa de mortalidad infantil (infant mortality for both sexes) con la esperanza de vida para ambos sexos (life expectancy at birth for both sexes) para el año 2020.

```python
# Importamos nuestro .csv al proyecto para poder trabajar
df = pd.read_csv('PopulationGrowth-F-MI.csv', compression=None, encoding= 'latin-1')
df
```

```python
# Para mayor practicidad y legibilidad, se modifican el nombre de las columnas marcadas
df.rename(columns={"Region/Country/Area":"Country","Unnamed: 1":"Region"}, inplace=True)
df
```

# Filtrado de datos (por año 2020)

```python
# Una vez tenemos listo el DF, empezaremos filtrando por el año a analizar la imformacion solicitada
df_2020 = df[df["Year"] == 2020]
df_2020
```

# Filtrar por series

```python
# Una vez que este nuestro filtro, pasaremos de demograficar las dos secciones que requeriremos para la correlacion de estos
mortalidad = df_2020[df_2020["Series"] == "Infant mortality for both sexes (per 1,000 live births)"]
esp_vida = df_2020[df_2020["Series"] == "Life expectancy at birth for both sexes (years)"]
```

# Seleccionar columnas relevantes (incluyendo los valores)

```python
mortalidad = mortalidad[["Country", "Value"]].rename(columns={"Value": "Mortalidad"})
esp_vida = esp_vida[["Country", "Value"]].rename(columns={"Value": "Esperanza_vida"})

# Posteriormente haremos la union de los dos conceptos en un DF nuevo
union = mortalidad.merge(esp_vida, on="Country")
```

# Graficar

```python
sns.scatterplot(data=union, x="Mortalidad", y ="Esperanza_vida", hue="Country")
plt.xlabel("Tasa de mortalidad infantil")
plt.ylabel("Esperanza de vida al nacer")
plt.title("Relación entre mortalidad infantil y esperanza de vida (2020)")

# Eliminamos los ticks  eje x y y para mayor claridad en la grafica 
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.tick_params(axis='y', which='both', left=False, right=False, labelleft=False)

plt.savefig("grafico_correlacion.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Tabla Pivot de los paises con mayor valor de esperanza de vida

```python
max_serie = df_2020.loc[df_2020.groupby("Region")["Value"].idxmax()]

# Ordenar de mayor a menor por la columna "Value"
max_tab_sorted = max_serie.sort_values(by="Value", ascending=False)
max_tab_sorted
```

```python
# Creamos la tabla pivot para un analisis de comparacion respecto a acada region que tenemos en la BD, pudiendo observa la media
# de la columna Value x Region
pivot_table = df_2020.pivot_table(index="Region", columns="Series", values="Value", aggfunc="mean")
pivot_table
```

```python
life_expectancy = df_2020[df_2020["Series"].str.contains("Life expectancy")]

plt.figure(figsize=(30, 6))
sns.barplot(data=life_expectancy, x="Region", y="Value")
plt.xticks(rotation=90)
plt.tick_params(axis='y', which='both', left=False, right=False, labelleft=False)
plt.title("Esperanza de vida al nacer por país (2020)")
plt.tight_layout()
plt.savefig("grafico_esp_x_region.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Asimismo, determine qué variables de esta base de datos se encuentran más correlacionadas en dicho año. Interprete sus resultados.

```python
# Primero empezaremos filtrando por el año 2020
df_2020 = df[df["Year"] == 2020].copy() # Crea una copia del DF para poder ser modificada, debido a que sin el ".copy()", 
                                        # estaremos modificando una vista y no una DF
```

```python
# Esto transformara las filas en columnas con indicadores
df_2020["Value"] = pd.to_numeric(df_2020["Value"], errors="coerce")  # Transforma los valores a tipo float y a los simbolos en Null
                                                    # De lo contrario no podremos avanzar ya que el tipo de valor que toma python es Int
# Creamos la tabla pivot
df_wide = df_2020.pivot_table(index="Country", columns="Series", values="Value")
df_wide
```

```python
# Eliminamos los valores que esten en nulo para evitar informacion erronea y poco presisa
df_wide = df_wide.dropna(axis=1, thresh=int(0.7 * len(df_wide)))
df_wide
```

```python
# Calculamos la matriz de correlacion
correlation = df_wide.corr(method="pearson") # Escojemos el metodo de "pearson"
```

```python
# Graficamos
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlación entre indicadores demográficos (2020)")
plt.savefig("grafico_calor_correlacional.png", dpi=300, bbox_inches='tight')
plt.show()
```
