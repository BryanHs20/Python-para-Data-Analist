#### CALENDAR

```python
"""  CREACION DE FUNCION DE NORMALIZAICON  """
def normalizar_week(valor):
    valor = str(valor).strip().lower()
    
    # Si ya está en formato semana-año
    if '-' in valor and valor.split('-')[0].isdigit():
        semana, año = valor.split('-')
        return f"{int(semana):02d}-{año}"
    
    # Si está en formato mes-año
    try:
        fecha = pd.to_datetime(valor, format='%b-%y', errors='coerce')
        if pd.isnull(fecha):
            fecha = pd.to_datetime(valor, format='%B-%y', errors='coerce')  # por si viene como "mes-año"
        if pd.notnull(fecha):
            semana = fecha.isocalendar().week
            año = fecha.year % 100  # para que coincida con "semana" en DF2
            return f"{semana:02d}-{año}"
    except:
        return np.nan

    return np.nan
```

```python
# Aplicamos la funcion al DF
df_sales["WEEK"] = df_sales["WEEK"].apply(normalizar_week)
```

## Union de los DataFrames
Haremos Merge sobre los DF mas reeelevantes, esto nos ayudara a conciliar de mejor manera la informacion, de acuerdo a un maquetado realizado previamente, las relaciones quedarian de la siguiente forma:


De forma explicada, quedaria de la siguiente forma

-CATEGORY- se relacionara con SEGMENT y PRODUCT mediante "CATEGORY_ID" a "CATEGORY"

-PRODUCT- se relacionara con SALES mediante "ITEM" A "ITEM_CODE"

-SALES- se relacionara con CALENDAR mediante "WEEK" A "WEEK"

### DF Para analisis de Productos por Categoria y/o Segmento
(df_pdt)

```python
"""  CREACION DE MERGE
este paso es un poco mas complicado de explicar, debido a que la relacion de SEGMENT y PRODUCT es mas especial, ya que se creara una 
columna en PRODUCT llamada SEGMENT
Para ello tendremos que validar que la informacion de estas 5 columnas, coincida en hambas tablas: 'CATEGORY', 'FORMAT', 'ATTR1', 'ATTR2', 'ATTR3'
Si conciden, se les asignara el valor de la Tabla Segemento a la nueva columna SEGMENT em PRODUCT

Esto nos dara nuestra primer seccion de DaraFrames
"""
```

```python
# Limpiamos las columnas
# Pasamos todo a mayusculas y sin espacios
cols = ['ID_CATEGORY', 'FORMAT', 'ATTR1', 'ATTR2', 'ATTR3']
for col in cols:
    df_segment[col] = df_segment[col].astype(str).str.strip().str.upper()
    df_pdt_clean[col] = df_pdt_clean[col].astype(str).str.strip().str.upper()
```

```python
df_pdt_clean = pd.merge(
    df_pdt_clean,
    df_segment[cols + ['SEGMENT']],  # solo las columnas necesarias
    on=cols,
    how='left'
)

# Renombrar la columna para que se llame SEGMENTO
df_pdt_clean.rename(columns={'SEGMENT': 'SEGMENTO'}, inplace=True)

df_pdt = df_pdt_clean.loc[:, ~df_pdt_clean.columns.duplicated()] # Borramos las columnas duplicadas de SEGMENTO
df_pdt
```

```python
# GUARDADO EN CSV Y EXCEL
df_pdt.to_csv("productos_por_cate.csv", index=False)
df_pdt.to_excel("productos_por_cate.xlsx", index=False)
```
