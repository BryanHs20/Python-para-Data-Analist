### Función auxiliar para obtener datos de pais

```python
# Esta función intentará obtener la Capital y la(s) Zona(s) Horaria(s)
# Usaremos caché para evitar consultas repetidas al objeto CountryInfo
# La función no es la lambda, pero será llamada por la lambda/apply.

# Diccionario de caché para almacenar resultados y evitar recalcular
country_data_cache = {} 

def get_country_details(country_name):
    """
    Intenta obtener la capital y las zonas horarias de un país.
    Utiliza un caché simple para países ya procesados.
    """
    if country_name in country_data_cache:
        return country_data_cache[country_name]
    
    try:
        # Inicializa CountryInfo con el nombre del país
        country = CountryInfo(country_name)
        
        # Obtener la capital. country.capital() devuelve un string.
        capital = country.capital() if country.capital() else 'N/A'
        
        # Obtener las zonas horarias. country.timezones() devuelve una lista.
        # Unimos las zonas horarias en un solo string, separadas por coma.
        timezones = ", ".join(country.timezones()) if country.timezones() else 'N/A'
        
        result = (capital, timezones)
    
    except KeyError:
        # Manejar países que countryinfo no reconoce
        result = ('Country Not Found', 'Country Not Found')
    except Exception:
        # Manejar otros errores
        result = ('Error', 'Error')

    # Almacenar el resultado en el caché antes de devolverlo
    country_data_cache[country_name] = result
    
    return result

# --- 3. Aplicación de la Lamba al DataFrame ---

# La función lambda será la encargada de llamar a la función auxiliar
# y desempaquetar la tupla resultante en las nuevas columnas.

# Aplicamos la función auxiliar a la columna 'Nationality'
# El método .apply() pasa cada valor de la columna a la lambda.
df_convert[['Capital', 'Timezones']] = df_convert['Nationality'].apply(
    lambda x: pd.Series(get_country_details(x))
)

# --- 4. Mostrar el resultado ---
```

```python
import pandas as pd
from datetime import datetime
from zoneinfo import ZoneInfo


# Lambda para obtener la capital
get_capital = lambda pais: df_convert.get(pais, {}).get("capital", "Desconocida")

# Lambda para obtener la zona horaria
get_zona = lambda pais: df_convert.get(pais, {}).get("zona_horaria", "UTC")

# Lambda para obtener la hora local según zona horaria
get_hora_local = lambda pais: datetime.now(
    ZoneInfo(df_convert.get(pais, {}).get("zona_horaria", "UTC"))
).strftime("%Y-%m-%d %H:%M:%S")

# Agregar las nuevas columnas al DataFrame
df["capital"] = df["Nationality"].apply(get_capital)
df["zona_horaria"] = df["Nationality"].apply(get_zona)
df["hora_local"] = df["Nationality"].apply(get_hora_local)

# Mostrar algunas filas de ejemplo
print(df[["Name", "Nationality", "capital", "zona_horaria", "hora_local"]].head())
```
