# Visualizacion con Seaborn

```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df
```

## Comparacion de graficos Pandas - Seaborn

```python
"""  Histograma con Pandas  """
df["math score"].plot.hist()
```

```python
"""  Histograma con Seaborn  """
import seaborn as sns
sns.distplot(df["math score"])
```

```python
"""  Histograma alternativo de Seaborn  """
sns.displot(df["math score"])
```

```python
sns.displot(df["math score"], kde = True, bins = 10)
```

```python
sns.distplot(df["math score"], hist = False)
```

```python
sns.distplot(df["math score"], hist = False, kde_kws = {"shade":True})
```

## Diagramas de dispersion

```python
"""  Creacion de diagrama con tecta de regresion  """
sns.regplot(x="reading score", y = "writing score", data = df)
```

```python
sns.lmplot(x="reading score", y="writing score", data=df)
```

```python
"""  Grafico de dispersion con regresion lineal por categoria de genero  """
sns.lmplot(x="reading score", y="writing score", data=df, hue = "gender")
```

```python
"""  Division de graficos  """
sns.lmplot(x="reading score", y="writing score", data=df, col = "gender")
```

```python
sns.lmplot(x="reading score", y="writing score", data=df, col = "race/ethnicity")
```
