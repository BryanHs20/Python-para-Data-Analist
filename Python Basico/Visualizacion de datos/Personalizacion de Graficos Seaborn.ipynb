```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
import seaborn as sns
"""  Tipos de estilos  """
for style in ["dark", "whitegrid","darkgrid","ticks"]:
    sns.set_style(style)
    sns.displot(df["reading score"])
```

```python
"""  Remocion de ejes  """
sns.set_style("ticks")
sns.displot(df["reading score"], kde=True)
#sns.despine(left=True) # Borra eje izquierdo
#sns.despine(bottom=True) # Borra eje inferior
sns.despine(left=True, bottom=True) # Borra ambos ejes
```

```python
"""  Combinacion de parametros con Matplotlib  """
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sns.distplot(df["writing score"], ax= ax)
ax.set(xlabel ="Score de escritura", ylabel = "Frecuencia relativa")
```

```python
"""  Truncamiento de escala de visualizacion  """
fig, ax = plt.subplots()
sns.distplot(df["writing score"], ax= ax)
ax.set(xlabel ="Score de escritura", ylabel = "Frecuencia relativa", xlim=(0,90), title="Histograma de frecuencias para puntajes de escritura")
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, ncols=2, sharey=True)
sns.distplot(df["writing score"], ax = ax0) # Histograma para todas las observaciones
sns.distplot(df.query('gender == "male"')["writing score"], ax = ax1) # Histograma para los hombres
ax0.set(xlabel="Puntaje general de escritura")
ax1.set(xlabel="Puntaje de esritura para hombres")
ax1.axvline(x = 75, label ="Objetivo", linestyle ="--") # Coloca una linea en el valor a asignar como referencia
ax1.legend()
```

```python
"""  Asignacion de colores  """
paleta=["deep", "muted", "pastel", "bright", "dark", "colorblind"]

for p in paleta:
    sns.set_palette(p)
    sns.displot(df["reading score"], kde=True)
```
