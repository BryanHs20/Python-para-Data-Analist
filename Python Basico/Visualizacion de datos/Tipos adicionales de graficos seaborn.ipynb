```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
"""  Creacion de un stripplot  """
import seaborn as sns
sns.stripplot(data = df, y = "writing score", x = "math score", jitter = True)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sns.stripplot(data= df, y="writing score", x="math score", jitter = True)
ax.axes.xaxis.set_ticks([])
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sns.stripplot(data= df, y="writing score", x="race/ethnicity", jitter = True, hue="race/ethnicity", palette="Set2")
plt.show()
```

```python
"""  Creacion de swarmplot  """
# Busca colocar loas observaciones de manera que no se empalmen
sns.swarmplot(data=df, y="writing score", x="race/ethnicity", hue="race/ethnicity", palette="Set2")
```

```python
"""  Creacion de boxplot  """
sns.boxplot(data=df, y="writing score", x="race/ethnicity", hue="race/ethnicity", palette="Set2")
```

```python
"""  Creacion de violinplot  """
sns.violinplot(data=df, y="writing score", x="race/ethnicity", hue="race/ethnicity", palette="Set2")
```

```python
"""  Crecion de lvplot (LetterValuePlot)  """
# Conviene para bases de datos grandes
sns.catplot(data=df, y="writing score", x="race/ethnicity", kind="boxen", hue="race/ethnicity", palette="Set2")
```

```python
"""  Creacion de barplot  """
sns.barplot(data=df, y="writing score", x="race/ethnicity", hue="race/ethnicity", palette="Set2")
```

```python
## Comparacion de grupos
sns.barplot(data=df, y="writing score", x="race/ethnicity", hue="lunch", palette="Set1")
```

```python
"""  Creacion de pointplot  """
sns.pointplot(data=df, y="writing score", x="race/ethnicity", hue="race/ethnicity", palette="Set2")
```

```python
sns.pointplot(data=df, y="writing score", x="race/ethnicity", hue="lunch", palette="Set1")
```

```python
"""  Creacion de countplot  """
sns.countplot(data=df, y="race/ethnicity", hue="lunch", palette="Set1")
```

```python
"""  Creacion de regplot (Grafico de regresion)  """
sns.regplot(data=df, y="writing score", x="reading score", marker = "+")
```

```python
"""  Creacion de residuales  """
sns.residplot(data=df, y="writing score", x="reading score")
```

```python
"""  Creacion de graficos con regresion con regplo y media por valor de x"""
# Util para enfatizar tendencias
import numpy as np
sns.regplot(data=df, y="writing score", x="reading score", x_estimator= np.mean)
```

```python
"""  Creacion de heatmap (mapa de calor)  """
sns.heatmap(df.corr())
```

```python
sns.heatmap(df.corr(), annot = True)
```
