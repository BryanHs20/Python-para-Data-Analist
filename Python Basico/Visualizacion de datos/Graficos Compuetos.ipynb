```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df
```

# Graficos Compuestos

```python
"""  Uso de FacetGrig  """
#FacetGrid es parte de la libreria de Seaborn

import seaborn as sns
graf = sns.FacetGrid(df, col = "race/ethnicity")
graf.map_dataframe(sns.boxplot, "math score", "race/ethnicity", order = ["group A", "group B", "group C", "group D", "group E"])
```

```python
"""  Uso de Factorplot  """
sns.catplot(x="math score",data = df, col="race/ethnicity", kind="box")
```

```python
"""  FacetGrid para graficos de dispersion  """
import matplotlib.pyplot as plt
graf= sns.FacetGrid(df, col = "race/ethnicity")
graf.map(plt.scatter, "reading score", "writing score")
```

```python
"""  Graficos tabulares de regresion  """
sns.lmplot(data=df, x="reading score", y="writing score", col="gender", row="race/ethnicity")
```

```python
"""  Creacion de una grafica conjunta  """
graf = sns.JointGrid(data=df, x="reading score", y="writing score")
graf.plot(sns.regplot, sns.distplot)
```

```python
sns.jointplot(data=df, x="reading score", y="writing score", kind="hex")
```

```python
"""  Filtrado de puntos a graficar  """
graf = sns.jointplot(data = df.query('gender == "male"'), x="reading score", y="writing score", kind="scatter")
```
