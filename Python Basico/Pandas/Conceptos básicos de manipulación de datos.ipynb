# Manejo de datos

## Introduccion

En datasets reales, es común que falten valores. Pandas los representa como NaN (Not a Number).
Podemos detectar, eliminar o rellenar estos valores.

- df.isnull() → Detecta valores nulos (True/False)
- df.isnull().sum() → Cuenta valores nulos por columna
- df.dropna() → Elimina filas con valores nulos
- df.fillna(valor) → Rellena valores nulos con un valor fijo
- df.fillna(df["col"].mean()) → Rellena con media, mediana, etc.
- Parámetros útiles:
- axis=0 → Filas (por defecto)
- axis=1 → Columnas
- subset=["col1", "col2"] → Solo en columnas específica

```python
#Ejemplo
df = pd.read_csv("olympic-medals.csv").isnull().sum()
df # Nulos
```

```python
#Elimina nulos
df = pd.read_csv("olympic-medals.csv").dropna()
df
```

```python
#Rellena nulos con 0
df = pd.read_csv("olympic-medals.csv").fillna(0)
df
```

# Eliminacion de duplicas

## Introduccion

A veces los datos contienen filas repetidas. Pandas permite detectarlas y eliminarlas.
🔑 Funciones clave
- df.duplicated() → Devuelve True si la fila está duplicada
- df.drop_duplicates() → Elimina duplicados
- Parámetros útiles:
- subset=["col1", "col2"] → Considera solo ciertas columnas
- keep="first" → Mantiene la primera ocurrencia
- keep="last" → Mantiene la última ocurrencia
- keep=False → Elimina todas las ocurrencias duplicadas

```python
# Detectar duplicados
print(df.duplicated().sum())

# Eliminar duplicados manteniendo la primera ocurrencia
df_sin_dups = df.drop_duplicates()

# Eliminar duplicados en la columna "ID"
df_sin_dups_id = df.drop_duplicates(subset=["ID"], keep="last")
```

# Union de DF

## Introduccion

Podemos combinar DataFrames de varias formas:
- Concatenación → Apilar uno encima de otro o lado a lado (pd.concat)
- Merge → Unir por columnas clave, como en SQL (pd.merge)
- Join → Similar a merge pero usando índices (df.join)
🔑 Funciones clave
- pd.concat([df1, df2], axis=0) → Une filas
- pd.concat([df1, df2], axis=1) → Une columnas
- pd.merge(df1, df2, on="columna") → Une por columna clave
- Tipos de merge:
- how="inner" → Solo coincidencias
- how="left" → Todo de la izquierda
- how="right" → Todo de la derecha
- how="outer" → Todo de ambos


```python
# Concatenar filas
df_total = pd.concat([df1, df2], axis=0)

# Merge por columna "ID"
df_merge = pd.merge(df1, df2, on="ID", how="inner")
```
