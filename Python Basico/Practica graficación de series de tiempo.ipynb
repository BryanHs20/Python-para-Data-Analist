```python
import matplotlib.pyplot as plt

fechas = df[df["Date"] >= "2015-01-01"]

fig,ax = plt.subplots()
ax.plot(fechas["Date"], fechas["Interpolated"])
plt.show
```

```python
import matplotlib.pyplot as plt

fechas = df[df["Date"] >= "2015-01-01"]

fig,ax = plt.subplots()
ax.plot(fechas["Date"], fechas["Interpolated"])

"""  Ajuste de titulo eje x  """
ax.axes.xaxis.set_ticks([])
plt.xlabel("Datos mensuales: 2015-01-01 - 2018-09-01")

plt.ylabel("Emisiones de CO2 promedio")
plt.show
```

```python
"""  Creacion de funcion que grafica series de tiempo  """
def grafica_series_tiempo(ejes, x, y, color, etiqueta_x, etiqueta_y):
    ejes.plot(x,y, color=color)
    ejes.set_xlabel(etiqueta_x)
    ejes.set_ylabel(etiqueta_y, color= color)
    ejes.tick_params('y', colors=color)
```

```python
fig,ax = plt.subplots()
grafica_series_tiempo(ax, fechas["Date"], fechas["Interpolated"], "blue", "Datos mensuales: 2015-01-01 - 2018-09-01", "CO2")

ax.axes.xaxis.set_ticks([])
plt.show()
```

```python
fig,ax = plt.subplots()
grafica_series_tiempo(ax, fechas["Date"], fechas["Interpolated"], "blue", "Datos mensuales: 2015-01-01 - 2018-09-01", "CO2 Interpolated")

ax2 = ax.twinx()

grafica_series_tiempo(ax2, fechas["Date"], fechas["Trend"], "red", "", "CO2 Trend")
ax.axes.xaxis.set_ticks([])
plt.show()
```
